```python
# Import Dependencies
import pandas as pd
```

```python
# Make a reference to the weekly data file path
csv_path = "weekly_data.csv"

# Import the csv file as a DataFrame
weekly = pd.read_csv(csv_path, index_col = 0)
weekly.head()
```

```python
html = weekly.to_html(index=False, na_rep='NaN', sparsify=None, index_names=True, justify=None, bold_rows=True, 
               classes="container-float", escape=True, show_dimensions=False, notebook=False)

```

```python
text_file = open("WeeklyData.html", "w")
text_file.write(html)
text_file.close()
```
