```python
import pandas as pd

data = pd.read_csv('..\..\DataFiles\sandp_data_final.csv')
```

```python
#have to transform this to weekly data by ticker symbol
import datetime

weeks = []
more_days = []

for i, row in data.iterrows():
    x = pd.to_datetime(row.date)
  
    more_days.append(x.date())
    week = x.isocalendar()[1]
    weeks.append(week)
    
```

```python
data['date'] = more_days


data['weekNo'] = weeks
data = data.sort_values(['stock', 'weekNo'], ascending=[True, True])

data
```

```python

prices = pd.read_csv('..\..\DataFiles\withPricesAndChange.csv')
prices = prices.sort_values(['Symbol', 'WeekNo'], ascending=[True, True])

some_days = []

for i, row in prices.iterrows():
    x = pd.to_datetime(row.Date)
    some_days.append(x.date())

prices['date'] = some_days
    
```

```python
prices['stock'] = prices['Symbol']
prices['weekNo'] = prices["WeekNo"]
prices
```

```python
#prices and data

# what we need is to get this price data in every cell where the date is the same as the other date 
 #mmkay lets do that
import numpy as np

pen = []
clo = []

dates = data['date'].unique()
tickz = data['stock'].unique()



    

    


df = pd.merge(data, prices[['date', 'stock','Opening', 'Closing', 'weekNo']], on=['stock', 'date'] , how='left')

df = df.drop_duplicates()

df
```

```python
df = df.sort_values(['stock', 'date','weekNo_x'], ascending=[True, True, True])


df['year'] = pd.DatetimeIndex(df['date']).year
df['week'] = df['weekNo_x']
df = df.fillna(method='ffill')
df = df[df['year'] > 2009]
a_df = df[df['stock'] == 'A']
```

```python
a_df = a_df[a_df['year'] == 2016]
a_df.sort_values(['year', 'week'])
```

```python
#now we need to add all this shit up to get the data we want to visualize. 
# we need a big ol dataframe ordered by ticker symbol and week number. it must have average Predicted scores and must also have the opening and closing prices for the week. 

tickers = []
predictions = []
opens = []
closes = []
week_num_num = []
yearz = []


stocks = df['stock'].unique()

years = df['year'].unique()

# big_dict = {
#     ticker = {
#         year = {
#             week = {

#             }
#         }
#     }

# }

big_dict = {}


for stock in stocks:
    #pulls all rows of a particular ticker
    temp_df = df[df['stock'] == stock]

    big_dict[stock] = {}

    for year in years:
        
        big_dict[stock][str(year)] = {}
        for i in range(1,53):

            big_dict[stock][str(year)][str(i)] = {}

            little_dict = {}

            weekly_df = temp_df[(temp_df['week'] == i) & (temp_df['year'] == year)]

            if(len(weekly_df.index) != 0):

                opens.append(weekly_df['Opening'].iloc[0])
                tickers.append(weekly_df['stock'].iloc[0])
                closes.append(weekly_df['Closing'].iloc[-1])
                week_num_num.append(i)
                yearz.append(year)
                
                predict_sum = 0
                for j, row in weekly_df.iterrows():
                    predict_sum += row['Predicted']
                
                avg_pred = predict_sum/len(weekly_df)
                predictions.append(predict_sum/len(weekly_df))

                little_dict = {"Ticker": weekly_df['stock'].iloc[0], "Open": float(weekly_df['Opening'].iloc[0]), "Close": float(weekly_df['Closing'].iloc[-1]), "Year" : str(year), "Week": int(i), "Prediction": float(avg_pred)}
                big_dict[stock][str(year)][str(i)] = little_dict







```

```python
d = {"Ticker": tickers, "Open": opens, "Close": closes, "Year" : yearz, "Week": week_num_num, "Prediction": predictions}

results_df = pd.DataFrame(d)
results_df

import json


big_json = json.dumps(big_dict)


with open('..\..\DataFiles\weekly_data.json', "w") as outfile:
    json.dump(big_json, outfile)

```
